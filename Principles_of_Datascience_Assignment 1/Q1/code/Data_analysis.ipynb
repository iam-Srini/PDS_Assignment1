```python
import pandas as pd
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('clean_data_file.csv')
print(data.to_string())
```

```python
sns.distplot(data['Frailty'])
plt.savefig('frailty.png', bbox_inches='tight', dpi=300)
```

```python
sns.distplot(data['Height'])
plt.savefig('height.png', bbox_inches='tight', dpi=300)
```

```python
sns.distplot(data['Weight'])
plt.savefig('weight.png', bbox_inches='tight', dpi=300)
```

```python
sns.distplot(data['Age'])
plt.savefig('age.png', bbox_inches='tight', dpi=300)
```

```python
sns.distplot(data['Grip_Strength'])
plt.savefig('Grip_Strength.png', bbox_inches='tight', dpi=300)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True, figsize =(15,3))

ax1.scatter(x=data['Height'],y=data['Frailty'])
ax1.set_title('Height and Frailty')
ax2.scatter(x=data['Weight'],y=data['Frailty'])
ax2.set_title('Weight and Frailty')
ax3.scatter(x=data['Age'],y=data['Frailty'])
ax3.set_title('Age and Frailty')

# Save the images
plt.savefig('frailty_scatter.png', bbox_inches='tight', dpi=300)


plt.show()
```
